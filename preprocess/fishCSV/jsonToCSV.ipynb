```
def makeRow(json, num, fishCode):
    top = json[num]['Label']['objects'][0]['bbox']['top']
    left = json[num]['Label']['objects'][0]['bbox']['left']
    height = json[num]['Label']['objects'][0]['bbox']['height']
    width = json[num]['Label']['objects'][0]['bbox']['width']
    ExternalID = json[num]['External ID']

    xmin = left
    xmax = left + width
    ymin = top
    ymax = top + height
    filename = ExternalID

    width = 224
    height = 224

    fishCode= "chu"

    row = []
    row.append(filename)
    row.append(width)
    row.append(height)
    row.append(fishCode)

    row.append(xmin)
    row.append(ymin)
    row.append(xmax)
    row.append(ymax)

    # print(row)

    return row
```

```
def makeDataFrame(fishCode):
  filename = fishCode+"Raw.json"
  resultFileName = fishCode+".csv"
  
  # 데이터 프레임 만들기
  import pandas as pd
  cols=['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
  df = pd.DataFrame(columns=cols)
  # print(df)

  # 파일 열고 json형태로 읽기
  json = openFile(filename)
  length = len(json)
  # print(length)

  for num in range(0, length):
    row = makeRow(json, num, fishCode)
    df.loc[len(df)] = row
    
    # print(df)

  df= df.sort_values(by='filename')
  df = df.reset_index(drop=True)
  return df
```
