# 사용방법
1. codeList 변수에 코드를 넣는다.
2. [코드] 폴더(폴더 이름이 "코드"이다)안에 [코드]Train.csv, codeTest.csv, codeFinalTest.csv를 넣는다.
 + 위의 csv 파일들은 "제이슨 csv로 만들기_동전.ipynb" 를 실행시켜 얻는다.
 + 제이슨 파일은 라벨박스 사이트에서 바운딩박스를 해서 얻는다.
3. 실행시킨다.

# 결과
* 현재폴더에 각 Train, Test, FinalTest끼리 합쳐진 csv 파일을 얻는다. 
* Train이 합쳐져서 -> train_labels.csv
* Test가 합쳐져서 -> eval_labels.csv
* FinalTest가 합쳐져서 -> final_test.csv

````
codeList=["ten",  "fft"]
cols=['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']

train = pd.DataFrame(columns=cols)
test = pd.DataFrame(columns=cols)
finalTest = pd.DataFrame(columns=cols)

outputList = [train, test, finalTest]

# print(outputList)
````

````
def openFile(code):
  path = "./"+code+"/"+code
  train = pd.read_csv(path+"Train.csv")
  test = pd.read_csv(path+"Test.csv")
  finalTest = pd.read_csv(path+"FinalTest.csv")
  dfList=[train, test, finalTest]
  return dfList
````

````
for code in codeList:
  codeList = openFile(code)
  for i in range(0, 3):
    outputList[i] = outputList[i].append(codeList[i])
````
